```python
import os
os.chdir('/home/jovyan/work')
import do_calibration
import reading
import init
```

```python
candidates = do_calibration.get_candidates(0)
```

```python
candidates_hash = {}
for entry in candidates:
    candidates_hash[entry.local_id] = entry

    
```

```python
def upsilonify_path(the_path):
    the_dir = os.listdir(the_path)
    the_dir.sort()
    results = {}
    for name in the_dir: # 'file' is a builtin type, 'name' is a less-ambiguous variable name.
        upsilon_entry = candidates_hash.get(reading.filename_to_star(name), None)
        if upsilon_entry is None:
            print("Skipping", name)
            continue
        #print(upsilon_entry.upsilon)
        new_name = "{}_{:.2f}_{}".format(upsilon_entry.upsilon[2], upsilon_entry.upsilon[1], name)
        print("Renaming {} to {}".format(name, new_name))
        os.rename(the_path+name, the_path+new_name)

def put_phase_in_the_back(the_path, suffix='_phase'):
    the_dir = os.listdir(the_path)
    the_dir.sort()
    results = {}
    for name in the_dir: # 'file' is a builtin type, 'name' is a less-ambiguous variable name.
        number = name[5:-4] # 0672
        ext = name[-4:] # .png
        new_name = number + suffix + ext
        print("Renaming {} to {}".format(name, new_name))
        os.rename(the_path+name, the_path+new_name)
        
def undo_upsilonify_path(the_path):
    the_dir = os.listdir(the_path)
    the_dir.sort()
    results = {}
    for name in the_dir: # 'file' is a builtin type, 'name' is a less-ambiguous variable name.
        new_name = name[7:]
        print("Renaming {} to {}".format(name, new_name))
        os.rename(the_path+name, the_path+new_name)

put_phase_in_the_back(init.phasedir)
#upsilonify_path(init.phasedir)
#undo_upsilonify_path(init.phasedir)
```
