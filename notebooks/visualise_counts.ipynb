```python
import pandas as pd
%matplotlib inline
df = pd.read_csv("counts.txt", delim_whitespace=True,
                         names=['count', 'file'], dtype={'count': int})
```

```python
import matplotlib.pyplot as plt
cf = df[df['count'] > 1102]
cf['count'].hist(bins=100, figsize=(20,10))
plt.xlabel("Nr of detections", fontsize=15)
plt.ylabel("Count",fontsize=15)
# plt.xscale('log')
```
